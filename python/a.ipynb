```python
from tkinter import *

def btn_switch():
    if btn.cget('text') == 'ON':
        btn.configure(text = 'OFF' , bg ='red')
    else:
        btn.configure(text='ON', bg='green')

main = Tk()
main.title("toggle")
main.geometry("300x200")
lbl = Label(main,text = "클릭하세요",font = "Arial 15")
lbl.pack()
btn = Button(main, text = "ON", command = btn_switch, bg='green')
btn.pack()


main.mainloop()
```

```python
from tkinter import *
from tkinter import ttk

main = Tk()
main.title("사칙연산 계산기")
frm = ttk.Frame(main, padding=10) #여백
frm.grid()

def plus() :
    result.set(a.get() + b.get())
    calc_progress.set(f"{a.get()} + {b.get()} =")

def minus() :
    result.set(a.get() - b.get())
    calc_progress.set(f"{a.get()} - {b.get()} =")

def multi() :
    result.set(a.get() * b.get())
    calc_progress.set(f"{a.get()} * {b.get()} =")

def divide() :
    if b.get() :
        result.set(a.get() / b.get())
        calc_progress.set(f"{a.get()} / {b.get()} =")
    else :
        result.set(0)
        calc_progress.set(f"0으로 나누기 불가")

def mod() :
    if b.get() :
        result.set(a.get() % b.get())
        calc_progress.set(f"{a.get()} % {b.get()} =")
    else :
        result.set(0)
        calc_progress.set(f"0으로 나누기 불가")

def reset() :
    a.set(0)
    b.set(0)
    calc_progress.set("")
    result.set(0)

ttk.Label(frm, text="정확한 사칙연산 계산기").grid(column=1, row=0)
ttk.Label(frm, text="계산할 두 수 입력 ->").grid(column=0, row=1)
a = IntVar(value=0)
ttk.Entry(frm, textvariable=a).grid(column=1, row=1)
b = IntVar(value=0)
ttk.Entry(frm, textvariable=b).grid(column=2, row=1)

plus_b = ttk.Button(frm, text="+", width = 20, command=plus)
minus_b = ttk.Button(frm, text="-", width = 20, command=minus)
multi_b = ttk.Button(frm, text="*", width = 20, command=multi)
divide_b = ttk.Button(frm, text="/", width = 20, command=divide)
mod_b = ttk.Button(frm, text="%", width = 20, command=mod)
reset_b = ttk.Button(frm, text="reset", width = 20, command=reset)

plus_b.grid(column=0, row=2)
minus_b.grid(column=1, row=2)
mod_b.grid(column=2, row=2)
multi_b.grid(column=0, row=3)
divide_b.grid(column=1, row=3)
reset_b.grid(column=2, row=3)

calc_progress = StringVar(value="")
result = IntVar(value=0)
ttk.Label(frm, text="계산결과는 : ").grid(column=0, row=4)
ttk.Label(frm, textvariable=calc_progress).grid(column=1, row=4)
ttk.Label(frm, textvariable=result).grid(column=2, row=4)

main.mainloop()
```
